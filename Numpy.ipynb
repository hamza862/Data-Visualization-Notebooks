```python
import numpy as np
import statsmodels.api as sm
import pylab as py
```

```python
# Standard normal distribution
data_points = np.random.normal(0,1,100)
data_points
```

```python
sm.qqplot(data_points, line = '45')
py.show()
```

```python
# Create normal distributed data
import numpy as np

# using np.lognormal( method)
gfg = np.random.lognormal(0, 1, 100)

print(gfg)
```

```python
sns.displot(gfg , kind ='kde')
```

```python
sm.qqplot(gfg, line = '45')
py.show()
```

```python
sm.qqplot(np.log(gfg), line = '45')
py.show()
```
